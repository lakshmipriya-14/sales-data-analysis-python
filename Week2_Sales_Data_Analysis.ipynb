```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(1)


products = ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Charger']
categories = ['Electronics', 'Accessories']


data = {
'Order_ID': range(1, 51),
'Order_Date': pd.date_range(start='2024-01-01', periods=50, freq='D'),
'Product': np.random.choice(products, 50),
'Category': np.random.choice(categories, 50),
'Quantity': np.random.randint(1, 6, 50),
'Price': np.random.randint(500, 50000, 50)
}


sales_df = pd.DataFrame(data)
sales_df.head()
```

```python
sales_df['Total_Sales'] = sales_df['Quantity'] * sales_df['Price']
sales_df.head()
```

```python
total_revenue = sales_df['Total_Sales'].sum()
total_orders = sales_df['Order_ID'].count()


print('Total Revenue:', total_revenue)
print('Total Orders:', total_orders)
```

```python
product_sales = sales_df.groupby('Product')['Total_Sales'].sum()
product_sales
```

```python
category_sales = sales_df.groupby('Category')['Total_Sales'].sum()
category_sales
```

```python
sales_df['Month'] = sales_df['Order_Date'].dt.month
monthly_sales = sales_df.groupby('Month')['Total_Sales'].sum()
monthly_sales
```

```python
sales_df.to_csv('week2_sales_analysis_output.csv', index=False)
```
